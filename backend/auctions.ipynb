```python
import pandas as pd
```

```python
df = pd.read_csv("inv2.csv")
df.head()
```

```python
def prune(keyword, df):
    if isinstance(keyword, str):
            keywords = [keyword]
    else:
        keywords = keyword
        pattern = '|'.join(keywords)

    filtered_df = df[~df['title'].str.contains(pattern, case=False, na=False)]
    return filtered_df


```

```python
pd.set_option('display.max_rows', None)
```

```python
ebay = df[['title', 'current_bid', 'bid_count', 'end_date']]
```

```python
ebay = prune(['45', '7"' ], ebay)
```
